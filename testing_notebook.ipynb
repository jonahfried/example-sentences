```python
from precision import *
import json
import pandas as pd
```

```python
wse = WikiSentenceExtractor()
```

```python
wse.run()
```

```python
example_sentences = {
    word:[json.loads(sentence)["string"] for (_, sentence) in sentences] 
    for (word, sentences) 
    in wse.example_sentences.items()
}
```

```python
example_sentences["david"]
```

```python
word_to_id = {word:id for (id, word) in enumerate(example_sentences.keys())}
```

```python
id_sentence_pairs = []
```

```python
for word, lst in example_sentences.items():
    for sentence in lst:
        id_sentence_pairs.append((word_to_id[word], word, sentence))
```

```python
pd.DataFrame(id_sentence_pairs)
```
